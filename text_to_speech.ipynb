```python
from gtts import gTTS   #gtts icinden gTTS i cekiyoruz
yazi="bana konus bana, burada millet sana ganalize mi olacak gebes gaplumbaa"  #okutmak istedigimiz yaziyi yaziyoruz
output=gTTS(text=yazi, lang="tr", slow=False)    #outputa gTTS icinde () text i atiyoruz, dili atiyoruz, slow olup olmadigini False olarak ayarliyoruz
output.save("seskaydi.mp3")    #outputumuzu icinde oldugumuz dosyaya kaydediyoruz.
```

```python
from IPython.display import Audio      #IPython.display icinden audioyu cekiyoruz
audio=Audio(filename="seskaydi.mp3")    #audioya kaydettigimiz ses dosyamizi atiyoruz
audio    #audioyu cagiriyoruz.
```
